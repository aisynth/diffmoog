This notebook is used to upload an out of domain audio synthesizer dataset and process it to be in a requested length

```python
import os
import glob
import librosa
import soundfile as sf

from scipy.io import wavfile

path = r"C:\Users\noamk\PycharmProjects\ai_synth\data\ARTURIA_dataset"
write_path = r"C:\Users\noamk\PycharmProjects\ai_synth\data\ARTURIA_1sec"
path2 = r"C:\Users\noamk\PycharmProjects\ai_synth\data\lfo_saw\val_nsynth\wav_files"
files = os.listdir(path)

# for filename in glob.glob(os.path.join(path2, '*.wav')):
#     sr, dat = wavfile.read(filename)
#     print(sr)
#     print(dat.dtype)
#     print(dat.shape)

for filename in glob.glob(os.path.join(path, '*.wav')):
    data, samplerate = librosa.load(filename, sr=16000, mono=True, duration=1)
    basename = os.path.basename(filename)
    sf.write(rf"{write_path}\{basename}.wav", data, samplerate, 'PCM_16')
```
