```python
import os, glob
import json
from pathlib import Path

import matplotlib.pyplot as plt
from scipy.io import wavfile
from PIL import Image

from IPython.display import Audio, display
```

```python
sounds_dir = Path('sound examples')


for preset_name in os.listdir(sounds_dir):

    print(f"*************************************************************")
    print(f"************************** {preset_name} ************************")
    print(f"*************************************************************")

    wav_file = glob.glob(os.path.join(sounds_dir, preset_name, '*.wav'))[0]
    params_file = glob.glob(os.path.join(sounds_dir, preset_name, '*.txt'))[0]
    spec_file = glob.glob(os.path.join(sounds_dir, preset_name, '*.png'))[0]

    with open(params_file, 'r') as f:
        params = f.read().split('\n')

    samplerate, signal = wavfile.read(wav_file)
    spec = Image.open(spec_file)

    audio = Audio(signal, rate=samplerate, autoplay=False)
    display(audio)

    fig, ax = plt.subplots(1, 1, figsize=(12, 3))
    ax.plot(signal)

    fig2, ax2 = plt.subplots(1, 1, figsize=(10, 3))
    ax2.imshow(spec)
    plt.show()
```
